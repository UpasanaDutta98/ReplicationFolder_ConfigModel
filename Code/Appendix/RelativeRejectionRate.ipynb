```python
import pickle
import matplotlib.pyplot as plt
import numpy as np
```

```python
pickle_in = open("../../Output/SwapsRejectionRateForSimpleNetworks/SwapRejectionRate_SimpleNetworkResults.pkl","rb")
results = pickle.load(pickle_in)
```

```python
plt.figure(figsize = (7, 4))
plt.scatter(results[1], results[7], c = "cyan", label = "Due to intro of multi-edge", s = 100, alpha = 0.2)
plt.xlabel("Number of edges (10\u2074), m", fontsize = 20, color = "black")
plt.ylabel("Probability of choosing two\nnon-adjacent edges, q", fontsize = 20, color = "black")
plt.xticks([0, 10000, 20000, 30000, 40000, 50000], [0, 1, 2, 3, 4, 5])
plt.xticks(fontsize = 20, color = "black")
plt.yticks(fontsize = 20, color = "black")
plt.ylim(0, 1.05)
plt.show()
```

```python
plt.figure(figsize = (8, 5))
plt.scatter(results[3], np.array(results[4])/np.array(results[5]), c = "orange", label = "Due to intro of multi-edge", s = 100, alpha = 0.4)
plt.xlabel("Density", fontsize = 20, color = "black")
plt.ylabel("Relative rejection rate, e / f", fontsize = 20, color = "black")
plt.xticks(fontsize = 20, color = "black")
plt.yticks(fontsize = 20, color = "black")
plt.xscale("log")
plt.yscale("log")
plt.ylim(1, 10000)
plt.show()
```
